## To find out the names of our team members

```python
import time, re
```

```python
cnt, all_names = 2, list()

while True:
    # break if the page don't have any upvote button in it
    if len(soup.find_all("button", class_=re.compile(r"btn btn-outline-success btn-sm"))) == 0: break    
    # Loop through all the root node/comment thread id values in a page
    for item in soup.find_all("button", class_=re.compile(r"btn btn-outline-success btn-sm")):
        # Extracting the comment id number
        digit = int(re.findall('[0-9]+', item['onclick'])[0])
        for each in soup.select("#comment"+str(digit))[0].find_all('font', class_='p-0'):
            all_names += [each.get_text().strip()]
    # Manually generating the next page web link to go to the page
    linkToOpen = f"https://www.appliedaicourse.com/course/11/Applied-Machine-learning-course?page={cnt}#comments"
    driver.get(linkToOpen)
    # Creating the Beatiful soup object
    soup = bs4.BeautifulSoup(driver.page_source, 'lxml')
    # Incrementing cnt for next page
    # using sleep to give time for chrome to navigate to the next page
    cnt += 1
```

```python
from collections import Counter 

x = Counter(all_names)
x.most_common()
```

```python
team_members = ['Applied AI Tech Admin', 'team aaic', 'Applied AI Course', 'Applied AI Course Team1', 
                'AppliedAI', 'AppliedAICourse Team', 'AppliedAI Course', 'AppliedAI Team', 'Applied_AI', 
                'appliedai course', 'AppliedAICourse', 'team']
```

```python
team_members = dict(zip(team_members, [1]*len(team_members)))
```

## To create the csv file

```python
linkToOpen = f"https://www.appliedaicourse.com/course/11/Applied-Machine-learning-course?page={1}#comments"
driver.get(linkToOpen)
# Creating the Beatiful soup object
soup = bs4.BeautifulSoup(driver.page_source, 'lxml')
```

```python
data_frame, all_ques_ans, all_names, cnt = list(), list(), list(), 2

while True:
    # break if the page don't have any upvote button in it
    if len(soup.find_all("button", class_=re.compile(r"btn btn-outline-success btn-sm"))) == 0: break    
    # Loop through all the root node/comment thread id values in a page
    for item in soup.find_all("button", class_=re.compile(r"btn btn-outline-success btn-sm")):
        # Extracting the comment id number
        digit = int(re.findall('[0-9]+', item['onclick'])[0])
        # Extracting the paragraphs of each comment thread indicated by the id value
        for each in soup.select("#comment"+str(digit))[0].find_all('p'):
            all_ques_ans += [each.get_text().strip()]
        # Extracting names: student names and tutor names as per the page
        for each in soup.select("#comment"+str(digit))[0].find_all('font', class_='p-0'):
            all_names += [each.get_text().strip()]
        for i in range(len(all_names)-1):
            # Allowing 2 consecutive(all_ques_ans[i], all_ques_ans[i+1]) paragraphs to exist <=> 
            # all_ques_ans[i] is from any random student and all_ques_ans[i+1] is from any one of our tutor.
            if not team_members.get(all_names[i]) and team_members.get(all_names[i+1]):
                # exporting th Q&A pair to the data_frame list(Later this list is used to create the Data Frame)
                data_frame += [(all_ques_ans[i], all_ques_ans[i+1])]
        all_ques_ans, all_names = list(), list()
    # Manually generating the next page web link to go to the page
    linkToOpen = f"https://www.appliedaicourse.com/course/11/Applied-Machine-learning-course?page={cnt}#comments"
    driver.get(linkToOpen)
    # Creating the Beatiful soup object
    soup = bs4.BeautifulSoup(driver.page_source, 'lxml')
    # Incrementing cnt for next page
    # using sleep to give time for chrome to navigate to the next page
    cnt += 1; # time.sleep(1)
```

```python
import pandas as pd
```

```python
# Constructing a DataFrame
columns = ['question', 'answer']
df = pd.DataFrame(data=data_frame, columns=columns)
df.head()
```

```python
df.to_csv('chat_bot_project.csv', index=False)
```
